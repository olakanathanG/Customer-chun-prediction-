```python
from sklearn.preprocessing import LabelEncoder

model=LabelEncoder()
df['gender']=model.fit_transform(df['gender'])

df.head()
```

```python
def countt(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column}: {df[column].value_counts()}')
```

```python
#fine number of distinctr values in each column
def find_uniqu(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column}: {df[column].nunique()}')
find_uniqu(df)
```

```python
df['Dependents']=model.fit_transform(df['Dependents'])
df['Partner']=model.fit_transform(df['Partner'])
df['PhoneService']=model.fit_transform(df['PhoneService'])
df['OnlineSecurity']=model.fit_transform(df['OnlineSecurity'])
df['OnlineBackup']=model.fit_transform(df['OnlineBackup'])
df['DeviceProtection']=model.fit_transform(df['DeviceProtection'])
df['TechSupport']=model.fit_transform(df['TechSupport'])
df['StreamingTV']=model.fit_transform(df['StreamingTV'])
df['StreamingMovies']=model.fit_transform(df['StreamingMovies'])
df['PaperlessBilling']=model.fit_transform(df['PaperlessBilling'])
df['Churn']=model.fit_transform(df['Churn'])
df['MultipleLines']=model.fit_transform(df['MultipleLines'])
df['InternetService']=model.fit_transform(df['InternetService'])
df['Contract']=model.fit_transform(df['Contract'])
df['PaymentMethod']=model.fit_transform(df['PaymentMethod'])
```

```python
# convert object into int
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
#remmove na
df.dropna(inplace=True)
```

```python
#convert float into int
df['MonthlyCharges']=df['MonthlyCharges'].astype(int)
df['TotalCharges']=df['TotalCharges'].astype(int)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Churn',axis=1),df['Churn'],test_size=0.2,random_state=42)
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(19,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```
